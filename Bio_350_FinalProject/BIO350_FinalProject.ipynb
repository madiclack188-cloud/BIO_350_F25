```python
categories = ['AverageSlope', 'Elevation', 'FireMortality', 'DistanceFromFireEdge', 'PreFireSODmortality']
values = [pds['AverageSlope'].mean(), pds['Elevation'].mean(), pds['FireMortality'].mean(), pds['DistanceFromFireEdge'].mean(), pds['PreFireSODmortality'].mean()]

# create bar graph of categores vs values
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph of Seedling Data Categories vs Values')
plt.show()
```

```python
# insert line of best fit
z = np.polyfit(range(len(values)), values, 1)
p = np.poly1d(z)
plt.scatter(categories, values)
plt.plot(categories, p(range(len(values))), "r--")
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Scatter Plot of Seedling Data Categories vs Values with Line of Best Fit')
plt.show()
```
