```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import plotdata_seedlings.csv as pds dataframe 
import pandas as pd
pds = pd.read_csv('plotdata_seedlings (1).csv')
```

```python
# plot first 10 line in pds
pds.head(10)
```

```python
categories = ['AverageSlope', 'Elevation', 'FireMortality', 'DistanceFromFireEdge', 'PreFireSODmortality']
values = [pds['AverageSlope'].mean(), pds['Elevation'].mean(), pds['FireMortality'].mean(), pds['DistanceFromFireEdge'].mean(), pds['PreFireSODmortality'].mean()]

# create bar graph of categores vs values
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph of Seedling Data Categories vs Values')
plt.show()
```

```python
# make the figure wider 
plt.figure(figsize=(10,6))
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph of Seedling Data Categories vs Values')
plt.show()
```

```python
# make the figuer even wider
plt.figure(figsize=(12,8))
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph of Seedling Data Categories vs Values')
plt.show()
```
