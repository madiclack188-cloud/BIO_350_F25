```python
import matplotlib.pyplot as plt
```

```python
def N_logistic(N_0, r, k, time):
    # Logistic growth model
    # Save N_values in a list
    N_values = [N_0]  
    # Save the initial population
    current_N = N_0
    for _ in range(time):
        current_N = current_N + r * current_N * (1 - current_N / k)
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N) 
    return N_values 
```

```python
#Run the function with 240 days with an initial population of 12
N_values = N_logistic(12, 0.6, 240, 35)
#Plot the results
```

```python
# Plot with days on the x-axis and population on the y-axis

days = list(range(36))  # Days from 0 to 35
plt.plot(days, N_values, marker='o')
plt.xlabel('Days')
plt.ylabel('Duckweed Population')
plt.title('Duckweed Population Growth Over Time')
plt.grid()
plt.show()
```
