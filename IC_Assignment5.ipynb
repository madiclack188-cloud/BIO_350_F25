```python
import matplotlib.pyplot as plt 
```

```python
# create a function that takes inital S, I, R, beta, death rate, and time, as inputs 
def SIR_model(S0, I0, R0, beta, d_rate, time):
    """
    Parameters:
    S0 = 999 (initional suseptible population size of threatened species
    I0 = 1 (initional Infected population size of threatened species
    R0 = 0 (initional removed population size of threatened species)
    d_rate = 0.5 (Death rate of threatened species)
    beta = 0.001 (probability of transmission per contact per day)
    time = 30 (time period of infection)
    """

    #import the variables given to the function
    #set up lists to hold S, I, R values over time
    Svalues = [S0]
    Ivalues = [I0]
    Rvalues = [R0]
    
    #set current values to initional values
    currentS = S0
    currentI = I0
    currentR = R0

    #for loop to go through time period 
    for i in range(time):
        #calculate next S, I, R values
        currentS = currentS - (beta * currentS * currentI) 
        currentI = currentI + (beta * currentS * currentI) - (d_rate * currentI)
        currentR = currentR + (d_rate * currentI)

        #append next values to lists and round to 3 decimal places
        Svalues.append(round(currentS, 3))
        Ivalues.append(round(currentI, 3))
        Rvalues.append(round(currentR, 3))
    return Svalues, Ivalues, Rvalues 
```

```python
#call the SIR_model function with the given in the docstring and save the output to variables
S0, I0, R0 = SIR_model (999, 1, 0, 0.001, 0.5, 30)
print(S0)
print(I0)
print(R0)
```

```python
#plot the results
plt.plot(S0, label='Susceptible')
plt.plot(I0, label='Infected')
plt.plot(R0, label='Removed')
plt.xlabel('Time (days)')
plt.ylabel('Population Size')
plt.title('SIR Model of Disease Spread in a Threatened Species')
plt.legend()
plt.show()
```
