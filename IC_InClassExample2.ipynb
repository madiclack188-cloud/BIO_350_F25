How are we importing and plotting population data.

```python
#Import "Population_Data.csv" into pandas DataFrame 
import pandas as pd
```

```python
#Import "Population_Data.csv" into pandas DataFrame
pop_data = pd.read_csv("Population_Data.csv")
pop_data.head()
```

```python
#Average the population from day 17 to day 60
avg_pop = pop_data[(pop_data["Days"] >= 17) & (pop_data["Days"] <= 60)]["Population"].mean()
print(f"Average population from day 17 to day 60: {avg_pop}")
```

```python
#plot the population data with "Days" on the x-axis and "Population" on the y-axis
import matplotlib.pyplot as plt

plt.plot(pop_data["Days"], pop_data["Population"], color='black', marker='o', linestyle='-', markersize=5)
plt.axhline(y=276, color='green', linestyle='--', linewidth=2)
plt.ylabel("Population")
plt.xlabel("Days")
plt.title("Population over Time")
plt.xlim(left=0)
plt.ylim(bottom=0)

# Set custom y-ticks to include 276 and label it
yticks = plt.yticks()[0].tolist()
if 276 not in yticks:
    yticks.append(276)
    yticks = sorted(yticks)
plt.yticks(yticks, [str(int(y)) if y != 276 else '276 (ref)' for y in yticks])

plt.show()
```
