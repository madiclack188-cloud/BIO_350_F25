```python
#create an algoritm that calcultes the population of 5 ages of a monarch butterfly population over 25 years
# the intital First year population is 250, the initial second year population is 550, the initial third year population is 650, the initial fourth year population is 750, and the initial fifth year population is 850
# on average 1 butterfly produces 7 offspring in the second year, 5 offspring in the third year, and 3 offspring in the fourth year
# the first year butterflies survive 25% to second year, 20% of second year butterflies survive to third year, 15% of third year butterflies survive to fourth year, and 10% of fourth year butterflies survive to fifth year
# create a defintion to calculate the return population of each age group and the total population for each year
def butterfly_population(years, initial_first_year, initial_second_year, initial_third_year, initial_fourth_year, initial_fifth_year): 
    #set explanation of parameters
    """
    Parameters
    years (int): number of years to simulate
    initial_first_year (int): initial population of first year butterflies
    initial_second_year (int): initial population of second year butterflies
    initial_third_year (int): initial population of third year butterflies
    initial_fourth_year (int): initial population of fourth year butterflies
    initial_fifth_year (int): initial population of fifth year butterflies
    Returns
    first_year_population (list): population of first year butterflies over the years
    second_year_population (list): population of second year butterflies over the years
    third_year_population (list): population of third year butterflies over the years
    fourth_year_population (list): population of fourth year butterflies over the years
    fifth_year_population (list): population of fifth year butterflies over the years
    total_population (list): total population of butterflies over the years
    """
    #Complete the alogoritm using matrix multiplication
    # create a matrix to represent the transition of population between years 
    transition_matrix = np.array([[0, 1, 5, 3, 0], 
                                  [0.25, 0, 0, 0, 0], 
                                  [0, 0.2, 0, 0, 0],
                                  [0, 0, 0.15, 0, 0],
                                  [0, 0, 0, 0.1, 0]])
    

    # initialize the population array with the initial populations
    inital_popultaiton_vector = np.array([initial_first_year, initial_second_year, initial_third_year, initial_fourth_year, initial_fifth_year])
    # create lists to store the population of each age group and total population over the years
    first_year_population = []
    second_year_population = []
    third_year_population = []
    fourth_year_population = []
    fifth_year_population = [] 
    total_population = []
    # append the current populations to the lists
    first_year_population.append(inital_popultaiton_vector[0])
    second_year_population.append(inital_popultaiton_vector[1])
    third_year_population.append(inital_popultaiton_vector[2])
    fourth_year_population.append(inital_popultaiton_vector[3])
    fifth_year_population.append(inital_popultaiton_vector[4])
    total_population.append(sum(inital_popultaiton_vector))
    # loop through the number of years to simulate
    for year in range(1, years + 1):
        # calculate the population for the next year using matrix multiplication
        inital_popultaiton_vector = np.dot(transition_matrix, inital_popultaiton_vector)
        # append the current populations to the lists
        first_year_population.append(inital_popultaiton_vector[0])
        second_year_population.append(inital_popultaiton_vector[1])
        third_year_population.append(inital_popultaiton_vector[2])
        fourth_year_population.append(inital_popultaiton_vector[3])
        fifth_year_population.append(inital_popultaiton_vector[4])
        total_population.append(sum(inital_popultaiton_vector))
    
    # return the population of each age group and the total population for each year
    return first_year_population, second_year_population, third_year_population, fourth_year_population, fifth_year_population, total_population
```

```python
# def butterfly_population(years, initial_first_year, initial_second_year, initial_third_year, initial_fourth_year, initial_fifth_year, total_population): 

# call butterfly_population function to simulate 50 years with initial populations of 250, 550, 650, 750, and 850 respectively
# return first_year_population, second_year_population, third_year_population, fourth_year_population, fifth_year_population, total_population
first_year_population, second_year_population, third_year_population, fourth_year_population, fifth_year_population,total_population = butterfly_population(25, 250, 550, 650, 750, 850)
```

```python
#plot the population of each age group and the total population over the years

plt.plot(first_year_population, label='First Year Population')
plt.plot(second_year_population, label='Second Year Population')
plt.plot(third_year_population, label='Third Year Population')
plt.plot(fourth_year_population, label='Fourth Year Population')
plt.plot(fifth_year_population, label='Fifth Year Population')
plt.plot(total_population, label='Total Population', linewidth=2, color='black')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Butterfly Population Over Time')
plt.legend()
plt.grid()
plt.show()
```
